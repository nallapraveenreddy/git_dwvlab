#18.Create a dataframe contain two columns name,age and transform the 'age' column using map() method and user-defined function to create a new column 'age_in_months'.

```python
df = pd.DataFrame({"name" : ['alex','bob','alice','john'],
                   'age':[20,30,25,40]})
def age_months(age):
  return age*12
df['age_in_months'] = df['age'].map(age_months)
df
```

#19.Create a dataframe contain two columns name,age and transform the 'age' column using map() method and dictionary to create a new column 'age_groups'.

```python
df = pd.DataFrame({"name" : ['surya','star','arya','varun'],
                   'age':[25,30,35,40]})
#create a dictionary 
age_group = {18:'18-24',25:'25-34',35:'35-44',45:'45+'}

# Apply map() method
df['age_groups'] = df['age'].map(age_group)
df
```
